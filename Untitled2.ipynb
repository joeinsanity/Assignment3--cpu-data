```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as npd
from sklearn.metrics import mean_squared_error
from sklearn.datasets import fetch_openml
```

```python
cpu = fetch_openml(data_id=227)
x_train,x_val,y_train,y_val = train_test_split(cpu.data,cpu.target,test_size=0.2, random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Input,Dense
nn=Sequential()
nn.add(Input((12,)))
nn.add(Dense(30,activation="sigmoid"))
nn.add(Dense(10,activation="sigmoid"))
nn.add(Dense(1))

nn.compile(optimizer="adam",loss="mse",metrics=["mse"])
history= nn.fit(x_train,y_train,epochs=200,validation_data=(x_val,y_val),verbose=0)

## differeent optimizer only
nn2=Sequential()
nn2.add(Input((12,)))
nn2.add(Dense(30,activation="sigmoid"))
nn2.add(Dense(10,activation="sigmoid"))
nn2.add(Dense(1))

nn2.compile(optimizer="SGD",loss="mse",metrics=["mse"])
history2= nn2.fit(x_train,y_train,epochs=200,validation_data=(x_val,y_val),verbose=0)


##more hidden layers
nn3=Sequential()
nn3.add(Input((12,)))
nn3.add(Dense(30,activation="sigmoid"))
nn3.add(Dense(25,activation="sigmoid"))
nn3.add(Dense(15,activation="sigmoid"))
nn3.add(Dense(10,activation="sigmoid"))
nn3.add(Dense(1))

nn3.compile(optimizer="adam",loss="mse",metrics=["mse"])
history3= nn3.fit(x_train,y_train,epochs=200,validation_data=(x_val,y_val),verbose=0)


## different hidden layer shape
nn4=Sequential()
nn4.add(Input((12,)))
nn4.add(Dense(50,activation="sigmoid"))
nn4.add(Dense(30,activation="sigmoid"))
nn4.add(Dense(1))

nn4.compile(optimizer="adam",loss="mse",metrics=["mse"])
history4= nn4.fit(x_train,y_train,epochs=200,validation_data=(x_val,y_val),verbose=0)
```

```python
##classification
clas= fetch_openml(data_id=1471)
```

```python
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
enc=OneHotEncoder(sparse=False)
tmp=[[x]for x in clas.target]
ohe_target=enc.fit_transform(tmp)
x_train,x_val,y_train,y_val=train_test_split(clas.data,ohe_target,test_size=0.2,random_state=0)

##1
nn=Sequential()
nn.add(Input(14,))
nn.add(Dense(20,activation="relu"))
nn.add(Dense(10,activation="relu"))
nn.add(Dense(2,activation="sigmoid"))
nn.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])

# Fit the model with one-hot encoded labels
history = nn.fit(x_train,y_train,epochs=200,validation_data=(x_val,y_val),verbose=0)

##2 more hidden layers
nn2=Sequential()
nn2.add(Input(14,))
nn2.add(Dense(30,activation="relu"))
nn2.add(Dense(25,activation="relu"))
nn2.add(Dense(20,activation="relu"))
nn2.add(Dense(10,activation="relu"))
nn2.add(Dense(2,activation="sigmoid"))
nn2.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])

# Fit the model with one-hot encoded labels
history2 = nn2.fit(x_train,y_train,epochs=200,validation_data=(x_val,y_val),verbose=0)


##3 same structure different optimizer
nn3=Sequential()
nn3.add(Input(14,))
nn3.add(Dense(20,activation="relu"))
nn3.add(Dense(10,activation="relu"))
nn3.add(Dense(2,activation="sigmoid"))
nn3.compile(optimizer="SGD",loss="categorical_crossentropy",metrics=["accuracy"])

# Fit the model with one-hot encoded labels
history3 = nn3.fit(x_train,y_train,epochs=200,validation_data=(x_val,y_val),verbose=0)



##4 differen hidden layer shape
nn4=Sequential()
nn4.add(Input(14,))
nn4.add(Dense(50,activation="relu"))
nn4.add(Dense(30,activation="relu"))
nn4.add(Dense(2,activation="sigmoid"))
nn4.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])

# Fit the model with one-hot encoded labels
history4 = nn4.fit(x_train,y_train,epochs=200,validation_data=(x_val,y_val),verbose=0)

```

```python
import matplotlib.pyplot as plt

# Function to plot training and validation accuracy
def plot_accuracy(history, title):
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

# Plotting for Model 1
plot_accuracy(history, "Model 1: 1 Hidden Layer, 20-10 nodes, Relu-Sigmoid")

# Plotting for Model 2
plot_accuracy(history2, "Model 2: 4 Hidden Layers, 30-25-20-10 nodes, Relu-Sigmoid")

# Plotting for Model 3
plot_accuracy(history3, "Model 3: 1 Hidden Layer, 20-10 nodes, Relu-Sigmoid, SGD Optimizer")

# Plotting for Model 4
plot_accuracy(history4, "Model 4: 3 Hidden Layers, 50-30 nodes, Relu-Sigmoid")
```
